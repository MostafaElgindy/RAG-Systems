```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

embeddings
```

```python
text = "Hello, I am learning about embeddings!"

embedding = embeddings.embed_query(text)
print(f"Text: {text}")
print(f"Len of embedding: {len(embedding)}")
print(f"Embedding: {embedding}")
```

```python
sentences = [
    "Hello, I am learning about embeddings!",
    "The weather is nice today.",
    "Embeddings are useful for many applications.",
    "Python is a great programming language.",
    "Artificial Intelligence is the future."
]

sentence_embeddings = embeddings.embed_documents(sentences)
for sentence, embed in zip(sentences, sentence_embeddings):
    print(f"Sentence: {sentence}")
    print(f"Len of embedding: {len(embed)}")
    print(f"Embedding: {embed}\n")
```

```python
print(sentence_embeddings[0])
print(embedding)
```

embed_documents and embed_query can produce different embeddings for the same text because they are optimized for different purposes
- embed_documents: used for embedding stored documents
- embed_query: used for embedding search queries
